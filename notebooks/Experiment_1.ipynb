```python
from ultralytics import YOLO

# Load pretrained COCO model 
model = YOLO("yolo11n.pt")

# Evaluate on your test set
metrics = model.val(data="mydata_capped.yaml", split="test")

# Print results
print(f"\n=== COCO Pretrained Baseline ===")
print(f"Precision: {metrics.box.mp:.3f}")
print(f"Recall: {metrics.box.mr:.3f}")
print(f"mAP@50: {metrics.box.map50:.3f}")
print(f"mAP@50-95: {metrics.box.map:.3f}")
```

```python
from ultralytics import YOLO

# -------------------------------------------------------
# LOAD PRETRAINED COCO MODEL
# -------------------------------------------------------
model = YOLO("yolo11n.pt")

# -------------------------------------------------------
# TRAIN ON YOUR REAL CCTV DATA
# Unfrozen backbone, high-resolution input, stable LR
# -------------------------------------------------------
model.train(
    data="mydata_capped.yaml",
    epochs=30,
    freeze=0,          # <-- fully unfrozen backbone (important)
    lr0=3e-4,          # <-- stable learning rate for small datasets
    imgsz=832,         # <-- boosts performance for small objects
    name="real_unfrozen_e30"
)

# -------------------------------------------------------
# EVALUATE ON REAL TEST SPLIT 
# This gives your final performance for comparison 
# with the COCO baseline model.
# -------------------------------------------------------
metrics = model.val(
    data="mydata_capped.yaml",
    split="test",
    plots=True,
    save_json=True
)


print("\n===== FINAL TEST METRICS =====")
print("mAP@0.50       :", metrics.box.map50)
print("mAP@0.50:0.95  :", metrics.box.map)
print("Precision      :", metrics.box.mp)
print("Recall         :", metrics.box.mr)
print("F1 Score       :", metrics.box.f1)
print("================================\n")
```
